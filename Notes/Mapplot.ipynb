Performing operations on the DF

Mapping Data
Data structures in Pandas are meant to be operated as a unit, meaning we can apply changes to the entire or a subset of DF


```python
# Works but not the best as we are manually looping each element
peso_price = []

for price in df['Price']:
    peso_price.append(price * 55)

#Re-assign the Price column to update it
df['Price'] = peso_price
df
```

```python
#Entire function to the collection - vectorized appraoch
df['Price'] = (df['Price'] * 55)
df #Advantage of Panda we can just call the entire column and apply a function to it
```

```python
#You can use columns to operate on each other.
df['Total'] = df['Price'] * df ['Units'] #Assigned a new colum
df
```

```python
#How we would normally turn the Product name uppercase

capitalized_product_names=[]

for product_name in df['Product']:
    capitalized_product_names.append(product_name.upper())
    
df['Product'] = capitalized_product_names
df
```

```python
#Using accessors to perform vectorized operations
# the .str accesor
df['Product'] = df['Product'].str.upper() #
df
```

Filter Data
Requires a condition that is met or not met
We can transform a column into a boolean column by comparing it to a value, then using the boolean column to index which rows in DataFrame

Essentially a helper column, allows to easily index

```python
df['Expensive'] = df['Price'] > 0.50
df.loc[2:5,['Price', 'Expensive']]
#Gives the value True or false
```

```python
df_expensive = df[df['Price'] > 0.50] #Creating a new data frame by filtering data
df_expensive #Using a helper columns makes it easier to verify if we filtered corrected/data is correct
```
